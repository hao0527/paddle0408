```python
import json
from pyecharts.charts import Pie
import pyecharts.options as opts
```

```python
# 读原始数据文件
datafile = 'data/statistics_data.json'
with open(datafile, 'r', encoding='UTF-8') as file:
    json_dict = json.loads(file.read())
    
# 分析各省份2月1日至今的累计确诊数据：'confirmedCount'
statistics__data = {}
for province in json_dict:
    statistics__data[province] = []
    for da in json_dict[province]:
        if da['dateId'] == 20200330:
            statistics__data[province].append(da['confirmedCount'])

#格式化
dataSequence=[]
for province in statistics__data:
    dataSequence.append((province,statistics__data[province][0]))

#对列表的第二个元素排序
dataSequence.sort(key=lambda x:x[1],reverse=True)

#绘饼图
Pie(init_opts=opts.InitOpts(width="600px", 
                            height="680px", 
                            bg_color="#2c343c"))\
.add(series_name="",
     data_pair=dataSequence,
     radius=["10%", "60%"],
     center=[300,420])\
.set_global_opts(title_opts=opts.TitleOpts(
                            title="累计确诊数据",
                            pos_left="center",
                            pos_top="20",
                            title_textstyle_opts=opts.TextStyleOpts(color="#fff")),
            legend_opts=opts.LegendOpts(is_show=False))\
.set_series_opts(label_opts=opts.LabelOpts(formatter="{b}: {c}"))\
.render_notebook()\
#.render("疫情分布饼图.html")
```
